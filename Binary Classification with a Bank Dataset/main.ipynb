# Random Forest

```python
%pip install xgboost
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report


# Prepare features and target (same as XGBoost, drop 'y'')
X_rf = df_another.drop(['y'], axis=1)
y_rf = df_another['y']

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_rf, y_rf)

# For evaluation, split a small validation set from training data
X_train_rf, X_val_rf, y_train_rf, y_val_rf = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)
rf.fit(X_train_rf, y_train_rf)

# Predict and evaluate
y_pred_rf = rf.predict(X_val_rf)
print("Random Forest Accuracy:", accuracy_score(y_val_rf, y_pred_rf))
print(classification_report(y_val_rf, y_pred_rf))
```

```python
# Predict probabilities for the test set using the trained Random Forest model
y_test_pred_proba = rf.predict_proba(df_test_another)[:, 1]

# Prepare the output DataFrame
output = pd.DataFrame({
    'id': range(750000, 750000 + len(df_test_another)),
    'y': y_test_pred_proba
})

# Save to CSV
output.to_csv('output.csv', index=False)
```

# Neural net

```python
from sklearn.neural_network import MLPClassifier

# Prepare features and target
X_nn = X_rf
y_nn = y_rf

# Split into train and validation sets
X_train_nn, X_val_nn, y_train_nn, y_val_nn = train_test_split(X_nn, y_nn, test_size=0.2, random_state=42)

# Define and train the neural network
mlp = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=10, random_state=42, verbose=True)
mlp.fit(X_train_nn, y_train_nn)

# Predict and evaluate
y_pred_nn = mlp.predict(X_val_nn)
print("Neural Net Accuracy:", accuracy_score(y_val_nn, y_pred_nn))
print(classification_report(y_val_nn, y_pred_nn))
```

```python
# Predict probabilities for the test set using the trained Random Forest model
y_test_pred_proba = np.round(mlp.predict_proba(df_test_another)[:, 1], 2)

# Prepare the output DataFrame
output = pd.DataFrame({
    'id': range(750000, 750000 + len(df_test_another)),
    'y': y_test_pred_proba
})

# Save to CSV
output.to_csv('output_NN.csv', index=False)
```

# XG Boost

```python
import xgboost as xgb
```

```python

# Prepare features and target for XGBoost
X_xgb = X_rf
y_xgb = y_rf

# Split into train and validation sets
X_train_xgb, X_val_xgb, y_train_xgb, y_val_xgb = train_test_split(X_xgb, y_xgb, test_size=0.2, random_state=42)

# Define and train the XGBoost model
xgb_clf = xgb.XGBClassifier(n_estimators=100, max_depth=6, learning_rate=0.1, random_state=42, use_label_encoder=False, eval_metric='logloss')
xgb_clf.fit(X_train_xgb, y_train_xgb)

# Predict and evaluate
y_pred_xgb = xgb_clf.predict(X_val_xgb)
print("XGBoost Accuracy:", accuracy_score(y_val_xgb, y_pred_xgb))
print(classification_report(y_val_xgb, y_pred_xgb))
```

```python
# Predict probabilities for the test set using the trained XGBoost model
y_test_pred_proba = xgb_clf.predict_proba(df_test_another)[:, 1]

# Prepare the output DataFrame
output = pd.DataFrame({
    'id': range(750000, 750000 + len(df_test_another)),
    'y': y_test_pred_proba
})

# Save to CSV
output.to_csv('output_xgboost.csv', index=False)
```
