```python
df_impute = df.copy()
df_impute['Stage_fear'] =df['Stage_fear'].fillna(df['Stage_fear'].mode()[0])
df_impute['Drained_after_socializing'] = df['Drained_after_socializing'].fillna(df['Drained_after_socializing'].mode()[0])
```

```python
df_impute.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

# df_impute = df.copy()
cols_to_impute = ['Time_spent_Alone', 'Social_event_attendance', 'Going_outside', 'Friends_circle_size', 'Post_frequency']
df_impute[cols_to_impute] = imputer.fit_transform(df[cols_to_impute])
```

```python
df_impute.isnull().sum()
```

```python
df_impute['Social_event_attendance'].value_counts()
```

```python
df_impute.select_dtypes(include=[np.number]).corr()
```

```python
df_impute['Anxiety_Interaction'] = df_impute['Drained_after_socializing'] * df_impute['Stage_fear']
```

```python
df_impute
```

```python
df_impute['Social_Conflict_Ratio'] = df_impute['Time_spent_Alone'] / (df_impute['Social_event_attendance'] + 1e-6)
```

```python
df_impute
```

```python
# Define bins and labels for 'Friends_circle_size'
bins = [0, 3, 10, np.inf]
labels = ['Small_Circle', 'Medium_Circle', 'Large_Circle']

df_impute['Friends_circle_category'] = pd.cut(df_impute['Friends_circle_size'], bins=bins, labels=labels, right=True, include_lowest=True)
```

```python
df_impute
```

```python
df_impute = pd.get_dummies(df_impute, columns=['Friends_circle_category'])
df_impute[['Friends_circle_category_Small_Circle', 'Friends_circle_category_Medium_Circle', 'Friends_circle_category_Large_Circle']] = df_impute[['Friends_circle_category_Small_Circle', 'Friends_circle_category_Medium_Circle', 'Friends_circle_category_Large_Circle']].astype(int)
```

```python
df_impute
```

```python
df_impute.drop(columns=['id'], inplace=True)
```

```python
df_impute
```

```python
X_train = df_impute[
    [
        'Time_spent_Alone',
        'Going_outside',
        'Post_frequency',
        'Anxiety_Interaction',
        'Social_Conflict_Ratio',
        'Friends_circle_category_Small_Circle',
        'Friends_circle_category_Medium_Circle',
        'Friends_circle_category_Large_Circle'
    ]
]

X_train
```

```python
X_train['Anxiety_Interaction'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

# Select columns to scale
cols_to_scale = [
    'Time_spent_Alone',
    'Going_outside',
    'Post_frequency',
    'Anxiety_Interaction',
    'Social_Conflict_Ratio'
]

scaler = StandardScaler()
X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])

X_train
```

```python
Y_train = df_impute['Personality']
```

```python
from sklearn.model_selection import train_test_split

X_tr, X_val, y_tr, y_val = train_test_split(X_train, Y_train, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Build the model
model = Sequential([
    Dense(32, activation='relu', input_shape=(X_tr.shape[1],)),
    Dense(16, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_tr, y_tr, epochs=20, batch_size=32, validation_data=(X_val, y_val))

# Evaluate
val_loss, val_acc = model.evaluate(X_val, y_val)
print(f"Validation Accuracy: {val_acc:.4f}")
```

```python


# Drop 'id' column if present
if 'id' in df_test.columns:
    df_test.drop(columns=['id'], inplace=True)

# Map categorical columns as in training
df_test['Stage_fear'] = df_test['Stage_fear'].map({'No': 0, 'Yes': 1})
df_test['Drained_after_socializing'] = df_test['Drained_after_socializing'].map({'No': 0, 'Yes': 1})

# Create interaction and ratio features
df_test['Anxiety_Interaction'] = df_test['Drained_after_socializing'] * df_test['Stage_fear']
df_test['Social_Conflict_Ratio'] = df_test['Time_spent_Alone'] / (df_test['Social_event_attendance'] + 1e-6)

# Bin and encode 'Friends_circle_size'
df_test['Friends_circle_category'] = pd.cut(
    df_test['Friends_circle_size'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)
df_test = pd.get_dummies(df_test, columns=['Friends_circle_category'])
for cat in labels:
    col = f'Friends_circle_category_{cat}'
    if col not in df_test.columns:
        df_test[col] = 0
    df_test[col] = df_test[col].astype(int)

# # Impute missing values
# df_test[cols_to_impute] = imputer.transform(df_test[cols_to_impute])

# Select features and scale
X_test = df_test[
    [
        'Time_spent_Alone',
        'Going_outside',
        'Post_frequency',
        'Anxiety_Interaction',
        'Social_Conflict_Ratio',
        'Friends_circle_category_Small_Circle',
        'Friends_circle_category_Medium_Circle',
        'Friends_circle_category_Large_Circle'
    ]
]
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
# Predict
y_pred = model.predict(X_test)
print(y_pred)
```
